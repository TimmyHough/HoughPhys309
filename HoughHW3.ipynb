```python
v2 = 0.99*con.c
v1 = np.linspace(0, 10, 50)
vr1 = (v1 +v2)/(1+(v1*v2/con.c))
vr2 = vr1*np.sin(con.pi/12)

plt.plot(v1, vr1, label='pure x-direction')
plt.plot(v1, vr2, label='ycomponant of outside cone')
plt.legend()
plt.xlabel('space station speed [m/s]')
plt.ylabel('speed measured by frank [m/s]')
```
